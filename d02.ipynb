```python
# min函数的不同调用方法 - 
# min, max, sum are built-in functions
min(1, 2, 3, 4, 5)
min([1, 2, 3, 4, 5])
min((1, 2, 3, 4, 5))
```

```python
help(print)
# print(*args, sep=' ', end='\n', file=None, flush=False)
# print is a built-in function
# *args is a tuple of positional arguments - 
```

```python
# 类型转换的内置函数 - casting or type conversion - 
# 都是函数 - 如 int(), float(), str(), list(), tuple(), set(), dict()
# int to float
num_int = 10
print('num_int',num_int)         # 10
num_float = float(num_int)
print('num_float:', num_float)   # 10.0

# float to int
gravity = 9.81
print(int(gravity))             # 9

# int to str
num_int = 10
print(num_int)                  # 10
num_str = str(num_int)
print(num_str)                  # '10'

# str to int or float
num_str = '10.6'
# print('num_int', int(num_str))      # 10 - invalid literal for int() with base 10: '10.6'
print('num_float', float(num_str))  # 10.6

# str to list
first_name = 'Asabeneh'
print(first_name)               # 'Asabeneh'
first_name_to_list = list(first_name)
print(first_name_to_list)            # ['A', 's', 'a', 'b', 'e', 'n', 'e', 'h']
```

```python
# 不能直接将字符串转换为整数，但可以先转换为浮点数，然后再转换为整数
print('num_int', int('10'))  # okay
print('num_float', float('10.0'))  # okay
print('num_int', int(float('10.0')))  # okay
print('num_int', int('10.0')) # ValueError: invalid literal for int() with base 10: '10.0'
```

```python
# calculate the area of a circle based on the radius
def area_of_circle(r):
    PI = 3.14
    area = PI * r ** 2 # 这儿的**是幂运算符的优先级高，不需要括号
    return area

# calculate the circumference of a circle
def circum_of_circle(r):
    PI = 3.14
    circumference = 2 * PI * r
    return circumference
```
